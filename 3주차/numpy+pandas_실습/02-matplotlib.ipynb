````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
xarr = np.random.rand(15).cumsum() # 앞에서부터 뒤로 더해가면서 값을 정하는 것
yarr = np.random.rand(15).cumsum()
xarr
````

````python
plt.figure(figsize=(4,3))   # figure 사이즈 조절
plt.plot(xarr)  # 파라미터를 하나만 넣어주면 xarr 값들을 y축의 값으로 사용하고, x 축은 데이터 개수
plt.show()
````

````python
plt.figure()  # 1개 이상의 Axes 를 가질 수 있음
plt.show()  # 실제 화면에 보이는 부분은 Axes (그래프) 임
````

````python
plt.figure()
plt.subplot()  # axes figure, subplot 을 생략해도 plot 사용시 내부적으로 실행함
# plt.plot(xarr)
plt.show()
````

### 1. 플롯을 여러 개 표시 (암시적 방법)

````python
plt.figure()
plt.subplot(2, 2, 1)  # 순서 중요
plt.plot(yarr)
plt.subplot(2, 2, 2)
plt.plot(xarr)
plt.show()
````

## 2. 명시적 작성

````python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax1.plot(xarr)
ax2.plot(yarr)
plt.show()
````

## 3. 미리 만들어 놓고 하기

````python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(xarr)
axes[1, 1].plot(yarr)
plt.show()
````

````python
axes
````

````python
fig, axes = plt.subplots(2, 2)
axes = axes.flatten()

for i in range(4):
    axes[i].plot(xarr)
plt.show()
````

````python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
````

````python
plt.figure(figsize=(8, 6))
plt.plot(xarr, label="korean", color="r", linestyle="--", marker="o")
plt.plot(yarr, label="english", color="g", linestyle="-.", marker="^")
plt.plot(zarr, label="japanese", color="c", linestyle=":", marker="*")
plt.title("This is title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.show()
````

## 산점도

````python
plt.figure(3, figsize=(8,6))
plt.scatter(xarr, yarr, marker='*', label='x', c='gold', s=10000)
plt.scatter(yarr, xarr, marker='o', label='o')
````

## boxplot

다섯가지 요약 수치
- 최소값
- 제 1사분위 수 (Q1)
- 제 2사분위 수 또는 중위수 (Q2)
- 제 3사분위 수 (Q3)
- 최대값



https://codedragon.tistory.com/7012

````python
x = np. random.normal(0, 2.0, 100)
y = np.random.normal(-3, 1.5, 500)
z = np.random.normal(1.2, 1.5, 1000)
````

````python
plt.boxplot([x,y,z])  # 집합의 범위와 중앙값 알 수 있음. 통계적으로 이상치(최소, 최대값을 넘어감)가 존재하는지 알 수 있음.
plt.show()
````
