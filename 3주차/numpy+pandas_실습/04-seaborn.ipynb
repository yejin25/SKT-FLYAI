```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.__version__
```

### Histogram

```python
tt = pd.read_csv("./datasets/titanic_train.csv")
tt
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=tt, x = "Age")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=tt, x = "Age", bins=10, kde = True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=tt, x = "Age", bins=10, hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=tt, x = "Age", bins=10, hue='Sex')
plt.show()
```

### displot

* figure-level plot
* matplotlib 과 연결되지 않으니 따로 설정해줘야함

```python
sns.displot(data=tt, x="Age")
plt.show()
```

```python
sns.displot(data=tt, x="Age", kde = True)
plt.show()
```

```python
sns.displot(data=tt, x="Age", kind='kde')
plt.show()
```

```python
sns.displot(data=tt, x="Age", col = 'Survived')
plt.show()
```

```python
sns.displot(data=tt, x="Age", col = 'Survived', hue='Sex')  # hue: Sex를 기준으로 그래프를 나눠 그려줌
plt.show()
```

### 막대 그래프

```python
sr = tt["Pclass"].value_counts()
```

```python
plt.bar(sr.index, sr.values)  # matplotlib
```

```python
sns.countplot(data=tt, x ='Pclass')  # seaborn
```

```python
sns.barplot(x='Pclass', y='Fare',data = tt)  # 검은 선 == 신뢰 구간, ci=False: 신뢰 구간 삭제
```

```python
sns.barplot(x='Pclass', y='Fare',data = tt, ci=False, hue='Sex')  # 각 카테고리의 비율
```

### Point plot

```python
sns.pointplot(x="Pclass", y='Fare', hue='Sex', data=tt)
```

### Boxplot

```python
sns.boxplot(x="Pclass", y="Age", data=tt, hue="Sex")
```

### Violinplot

- 인구 피라미드

```python
sns.violinplot(x="Pclass", y="Age", data=tt)
```

```python
sns.violinplot(x="Pclass", y="Age", hue="Sex", data=tt)
```

```python
sns.violinplot(x="Pclass", y="Age", hue="Sex", split=True, data=tt)
```

### Countplot

- 데이터 분석시에 많이 사용하는 그래프

```python
sns.countplot(x='Embarked', data=tt)
```

```python
sns.countplot(x="Sex", data=tt)
```

```python
sns.countplot(x="Survived", data=tt)
```

### Heatmap

```python
data = np.random.rand(4, 5)
data.shape
```

```python
sns.heatmap(data, annot=True, cmap="summer") # 밝을 수록 1에 가까운 값, 어둘울 수록 0에 가까운 값 | annot -> 값 표시 | cmap -> 색상 변경
```

### 산점도 Scatter plot

x축의 값과 y축의 값의 관계를 확인

```python
tips = sns.load_dataset('tips')
type(tips)
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)  # 상관관계가 있음
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

### Pair plot

- 분류 알고리즘
- 숫자 데이터가 많아야 제대로 그려짐
- 속성 간의 관계 파악 위함, 전체적으로 데이터를 보기 위함

```python
df = sns.load_dataset('penguins')  # 펭귄 종류 맞추는 데이터
df.head()
```

```python
df['species'].unique()
```

```python
sns.pairplot(df, hue='species')
```
